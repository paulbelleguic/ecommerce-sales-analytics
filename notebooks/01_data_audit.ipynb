## Objectif
Comprendre la structure du dataset Olist avant le nettoyage.

## Questions à traiter
- Quelles sont les tables disponibles ?
- Quelle est la granularité de chaque table ?
- Quelles colonnes contiennent des valeurs manquantes ?
- Quelles colonnes sont des dates ?
- Y a-t-il des doublons ?


```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data/raw")
DATA_DIR
```

```python
files = {
    "customers": "olist_customers_dataset.csv",
    "orders": "olist_orders_dataset.csv",
    "order_items": "olist_order_items_dataset.csv",
    "payments": "olist_order_payments_dataset.csv",
    "products": "olist_products_dataset.csv",
    "sellers": "olist_sellers_dataset.csv",
    "reviews": "olist_order_reviews_dataset.csv",
}

dfs = {}

for name, file_name in files.items():
    file_path = DATA_DIR / file_name
    dfs[name] = pd.read_csv(file_path)
    print(f"{name}: {dfs[name].shape}")
```

```python
def audit_table(df: pd.DataFrame, name: str) -> pd.DataFrame:
    summary = pd.DataFrame({
        "column": df.columns,
        "dtype": df.dtypes.astype(str).values,
        "missing_count": df.isna().sum().values,
        "missing_pct": (df.isna().mean() * 100).round(2).values,
        "n_unique": df.nunique(dropna=True).values
    })

    print(f"\n=== TABLE: {name} ===")
    print(f"shape: {df.shape}")
    print(f"duplicates: {df.duplicated().sum()}")
    display(df.head())
    display(summary)

    return summary
```

```python
audit_results = {}

for name, df in dfs.items():
    audit_results[name] = audit_table(df, name)
```

## Synthèse de l’audit

### Granularité
- `orders` : 1 ligne par commande
- `order_items` : 1 ligne par produit dans une commande
- `payments` : 1 ligne par paiement
- `products` : 1 ligne par produit
- `sellers` : 1 ligne par vendeur
- `reviews` : 1 ligne par avis, mais `order_id` n’est pas unique
- `customers` : 1 ligne par `customer_id`

### Colonnes date à convertir
- orders: `order_purchase_timestamp`, `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`, `order_estimated_delivery_date`
- order_items: `shipping_limit_date`
- reviews: `review_creation_date`, `review_answer_timestamp`

### Valeurs manquantes principales
- `orders`: plusieurs dates manquantes
- `products`: informations produit manquantes sur certaines lignes
- `reviews`: beaucoup de commentaires manquants

### Risques
- duplication des ventes lors des jointures
- confusion entre granularité commande et item
- mauvaise interprétation des valeurs manquantes



## Audit initial des données

Dans cette première étape, j’ai réalisé un audit initial du dataset e-commerce Olist afin de comprendre sa structure avant de commencer le nettoyage, l’analyse business et la modélisation.

J’ai d’abord chargé les principales tables du dataset dans des DataFrames Pandas à partir des fichiers CSV stockés dans le dossier `data/raw`. Cette étape m’a permis de vérifier que les données étaient bien accessibles et de mesurer la taille de chaque table à l’aide de leur dimension (`shape`), c’est-à-dire le nombre de lignes et de colonnes.

J’ai ensuite construit une fonction d’audit pour examiner chaque table de manière systématique. Pour chaque jeu de données, j’ai observé les colonnes disponibles, les types de variables, le nombre de valeurs manquantes, le nombre de valeurs uniques, les éventuels doublons ainsi qu’un aperçu des premières lignes.

Cette analyse a montré que le dataset est composé de plusieurs tables relationnelles avec des granularités différentes. La table `orders` semble être la table centrale car elle représente les commandes, tandis que `order_items` décrit les produits présents dans chaque commande et `payments` les paiements associés. Cette différence de granularité est essentielle, car une mauvaise jointure entre les tables pourrait entraîner une duplication artificielle des ventes.

L’audit a également permis d’identifier plusieurs colonnes de dates actuellement stockées sous forme de texte, qui devront être converties en format `datetime` pour permettre des analyses temporelles. Enfin, des valeurs manquantes ont été détectées dans certaines tables, notamment dans `orders`, `products` et `reviews`. Ces absences ne correspondent pas toutes à des erreurs de qualité : certaines peuvent s’expliquer par le contexte métier, par exemple l’absence de commentaire texte dans un avis client.

Cette étape d’audit constitue une base importante pour la suite du projet, car elle permet de préparer un nettoyage cohérent, de sécuriser les futures jointures SQL et Python, et d’éviter des erreurs d’interprétation dans l’analyse et la prédiction des ventes.

## Etape 02 - Nettoyage initial et conversion des données

Dans cette deuxième étape, l’objectif était de préparer les données pour les futures analyses business et pour la construction d’une table analytique exploitable. Après avoir identifié la structure générale du dataset lors de l’audit initial, il était nécessaire de réaliser un premier nettoyage technique sur les tables principales.

La première action a consisté à repérer les colonnes contenant des informations temporelles dans les tables `orders`, `order_items` et `reviews`. Lors du chargement initial, ces colonnes étaient interprétées comme des chaînes de caractères. Elles ont donc été converties au format `datetime` avec la fonction `pd.to_datetime()`. Le paramètre `errors="coerce"` a été utilisé afin de transformer les éventuelles valeurs invalides en valeurs manquantes temporelles (`NaT`) plutôt que de provoquer une erreur bloquante. Cette étape est essentielle pour permettre des analyses temporelles fiables, comme l’étude des ventes par mois, par jour ou par heure.

Ensuite, une analyse de la variable `order_status` a été réalisée afin de mieux comprendre le cycle de vie des commandes. Cette vérification a montré que la grande majorité des commandes sont au statut `delivered`, tandis qu’une partie plus faible correspond à des commandes annulées, indisponibles, en cours de traitement ou simplement créées. Cette observation est importante car toutes les commandes ne représentent pas nécessairement une vente finalisée. Il faudra donc tenir compte de ce statut lors des futures analyses.

Une vérification des identifiants principaux a également été effectuée. Les colonnes `order_id`, `customer_id`, `product_id` et `seller_id` se sont révélées uniques dans leurs tables respectives, ce qui confirme leur rôle d’identifiants fiables pour les jointures. En revanche, `review_id` n’est pas unique dans la table `reviews`, ce qui indique qu’il faudra manipuler cette table avec précaution.

L’étape suivante a consisté à comparer les identifiants de commandes présents dans plusieurs tables afin de détecter d’éventuels écarts. Cette vérification a montré que certaines commandes n’ont pas de ligne correspondante dans `order_items`, qu’une commande ne possède pas de paiement enregistré, et qu’un certain nombre de commandes n’ont pas d’avis associé. Ces résultats montrent que toutes les commandes ne sont pas systématiquement représentées dans toutes les tables, ce qui devra être pris en compte lors de la construction de la table analytique.

Enfin, une copie de travail de la table `orders` a été créée sous le nom `orders_clean`. Cette copie permet de transformer les données sans modifier directement la table brute d’origine. Des variables temporelles dérivées ont ensuite été ajoutées, comme l’année, le mois, le jour, l’heure d’achat et le jour de la semaine. Ces nouvelles variables seront utiles à la fois pour l’analyse exploratoire et pour de futurs modèles de machine learning.

Cette étape de nettoyage initial a donc permis de sécuriser les données, de préparer les futures jointures et de poser les bases d’une analyse temporelle des ventes.


## 03 - Build the Analytical Sales Table

```python
order_items_agg = (
    dfs["order_items"]
    .groupby("order_id", as_index=False)
    .agg(
        n_items=("order_item_id", "count"),
        n_unique_products=("product_id", "nunique"),
        n_unique_sellers=("seller_id", "nunique"),
        total_price=("price", "sum"),
        total_freight=("freight_value", "sum"),
    )
)

order_items_agg.head()
```

```python
payments_agg = (
    dfs["payments"]
    .groupby("order_id", as_index=False)
    .agg(
        n_payments=("payment_sequential", "count"),
        payment_value_total=("payment_value", "sum"),
        payment_installments_max=("payment_installments", "max"),
    )
)

payments_agg.head()
```

```python
customers_selected = dfs["customers"][
    ["customer_id", "customer_unique_id", "customer_city", "customer_state"]
].copy()

customers_selected.head()
```

```python
reviews_agg = (
    dfs["reviews"]
    .groupby("order_id", as_index=False)
    .agg(
        review_score_mean=("review_score", "mean"),
        review_count=("review_id", "count"),
    )
)

reviews_agg.head()
```

```python
sales_analytics = (
    orders_clean
    .merge(order_items_agg, on="order_id", how="left")
    .merge(payments_agg, on="order_id", how="left")
    .merge(customers_selected, on="customer_id", how="left")
    .merge(reviews_agg, on="order_id", how="left")
)

sales_analytics.head()
```

```python
print("sales_analytics shape:", sales_analytics.shape)
print("Unique order_id:", sales_analytics["order_id"].is_unique)
sales_analytics.isna().sum().sort_values(ascending=False).head(15)
```

```python
sales_analytics["delivery_delay_days"] = (
    sales_analytics["order_delivered_customer_date"] - sales_analytics["order_purchase_timestamp"]
).dt.days

sales_analytics["estimated_vs_actual_diff_days"] = (
    sales_analytics["order_delivered_customer_date"] - sales_analytics["order_estimated_delivery_date"]
).dt.days

sales_analytics["total_order_value"] = (
    sales_analytics["total_price"] + sales_analytics["total_freight"]
)

sales_analytics[
    [
        "order_id",
        "total_price",
        "total_freight",
        "total_order_value",
        "delivery_delay_days",
        "estimated_vs_actual_diff_days",
    ]
].head()
```

```python
delivered_sales = sales_analytics[sales_analytics["order_status"] == "delivered"].copy()

print("delivered_sales shape:", delivered_sales.shape)
print("Unique order_id:", delivered_sales["order_id"].is_unique)
```

## Étape 3 - Construction de la table analytique

L’objectif de cette étape était de construire une table analytique au niveau commande afin de disposer d’une base unique pour les analyses business, les requêtes SQL, le dashboard et le futur modèle de machine learning.

Pour cela, la table `orders_clean` a été utilisée comme table centrale, car elle représente les commandes et constitue l’unité d’analyse la plus adaptée au projet. Les tables `order_items`, `payments` et `reviews` ont d’abord été agrégées au niveau `order_id` afin d’éviter toute duplication des lignes lors des jointures. Cette précaution est essentielle car ces tables possèdent une granularité plus fine que la table des commandes.

L’agrégation de `order_items` a permis de calculer le nombre d’articles, le nombre de produits distincts, le nombre de vendeurs distincts, le montant total des produits et le total des frais de port pour chaque commande. L’agrégation de `payments` a permis de résumer le nombre de paiements, le montant total payé et le nombre maximal d’échéances. La table `customers` a été simplifiée pour conserver uniquement les informations utiles sur les clients, tandis que la table `reviews` a été agrégée par commande afin d’obtenir une note moyenne et un nombre d’avis.

L’ensemble de ces informations a ensuite été fusionné dans une table finale appelée `sales_analytics` à l’aide de jointures de type `left`, afin de conserver toutes les commandes, même lorsque certaines informations étaient absentes dans les tables secondaires.

La table obtenue contient une ligne par commande, avec un identifiant `order_id` toujours unique, ce qui confirme que la granularité a été préservée. Enfin, plusieurs variables métier ont été créées, notamment la valeur totale de la commande, le délai de livraison et l’écart entre la date de livraison réelle et la date estimée. Cette table analytique constitue désormais la base de travail principale pour les analyses business et la modélisation.
