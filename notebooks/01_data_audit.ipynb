# 01 - Data Audit

## Objectif
Comprendre la structure du dataset Olist avant le nettoyage.

## Questions à traiter
- Quelles sont les tables disponibles ?
- Quelle est la granularité de chaque table ?
- Quelles colonnes contiennent des valeurs manquantes ?
- Quelles colonnes sont des dates ?
- Y a-t-il des doublons ?


```python
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data/raw")
DATA_DIR
```

```python
files = {
    "customers": "olist_customers_dataset.csv",
    "orders": "olist_orders_dataset.csv",
    "order_items": "olist_order_items_dataset.csv",
    "payments": "olist_order_payments_dataset.csv",
    "products": "olist_products_dataset.csv",
    "sellers": "olist_sellers_dataset.csv",
    "reviews": "olist_order_reviews_dataset.csv",
}

dfs = {}

for name, file_name in files.items():
    file_path = DATA_DIR / file_name
    dfs[name] = pd.read_csv(file_path)
    print(f"{name}: {dfs[name].shape}")
```

```python
def audit_table(df: pd.DataFrame, name: str) -> pd.DataFrame:
    summary = pd.DataFrame({
        "column": df.columns,
        "dtype": df.dtypes.astype(str).values,
        "missing_count": df.isna().sum().values,
        "missing_pct": (df.isna().mean() * 100).round(2).values,
        "n_unique": df.nunique(dropna=True).values
    })

    print(f"\n=== TABLE: {name} ===")
    print(f"shape: {df.shape}")
    print(f"duplicates: {df.duplicated().sum()}")
    display(df.head())
    display(summary)

    return summary
```

```python
audit_results = {}

for name, df in dfs.items():
    audit_results[name] = audit_table(df, name)
```

## Premières observations

- `orders` semble être au niveau commande
- `order_items` semble être au niveau ligne de commande
- certaines colonnes contiennent des valeurs manquantes
- plusieurs colonnes de dates devront être converties au format datetime
- il faudra faire attention aux jointures pour éviter de dupliquer les ventes
